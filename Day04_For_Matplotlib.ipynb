```python
import matplotlib.pyplot as plt  #Import Matplotlib
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Create Sine Wave  with title 
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine Value")
plt.title("Sine Waves View")
plt.show()
```

```python
# Create Cosine Wave  with title 
plt.plot(x,z)
plt.xlabel("Angle")
plt.ylabel("Cosine Value")
plt.title("Cosine Waves View")
plt.show()
```

```python
# Creating A Parabola 
x = np.linspace(-30,30,30)
y = x**2
plt.plot(x,y)
plt.title("Parabola Shape")
plt.show()
```

```python
#Creating Parabola with colour symbol 

plt.plot(x,y,"r*")
plt.show()
```

```python
plt.plot(x,y,"g p")
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, np.sin(x),"r")
plt.plot(x,np.cos(x), "g*")
plt.show()
```

#Bar Plot Create 

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
language = ["English", "Bangla" , "Chaines" ,"Korian" , "Spanish" ,"French"]
People = [110 , 75 , 90 , 150 ,55 ,100]
ax.bar(language,People)
plt.xlabel("Language")
plt.ylabel("Number Of People")
plt.show()
```

# Create Pie Chart

```python
fig1 = plt.figure()
ax=fig1.add_axes([0,0,2,2])
language = ["English", "Bangla" , "Chaines" ,"Korian" , "Spanish" ,"French"]
People = [110 , 75 , 90 , 150 ,55 ,100]
ax.pie(People,labels=language,autopct="% 1.1f%%")
plt.show()
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
Country = ["Bangladesh","India","Pakisthan","Japan","Singapur"]
Affected = [1260,2650,765,460,1950]
ax.pie(Affected,labels=Country,autopct="%1.1f%%",radius=2)
plt.show()
```

#Scatter Plot

```python
x = np.linspace(0,10,50)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color="r")
plt.show()
```

# 3D Scatter Plor

```python
fig4 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
